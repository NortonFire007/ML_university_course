```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import string
import re
```

```python
x = np.linspace(0, 5, 100)
```

```python
# Y(x) = x ^ (1 / 2) * sin(10 * x), x = [0...5]
y = np.sqrt(x) * np.sin(10 * x)
```

```python
plt.plot(x, y, label=r'$Y(x) = \sqrt{x} \cdot \sin(10x)$')
plt.xlabel('x')
plt.ylabel('Y(x)')
plt.title('Графік функції Y(x) = sqrt(x) * sin(10x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
with open('text 2.txt', 'r', encoding='utf-8') as file:
    text = file.read().lower()

letters = [char for char in text if char in string.ascii_lowercase]

letter_counts = Counter(letters)

labels, values = zip(*letter_counts.items())

plt.bar(labels, values)
plt.xlabel('Літери')
plt.ylabel('Частота')
plt.title('Гістограма частоти появи літер у тексті')
plt.show()
```

```python
sentences = re.split(r'(?<=[.!?]) +', text)
```

```python
declarative_count = 0
question_count = 0
exclamation_count = 0
ellipsis_count = 0
```

```python
for sentence in sentences:
    match sentence:
        case s if s.endswith('...'):
            ellipsis_count += 1
        case s if s.endswith('?'):
            question_count += 1
        case s if s.endswith('!'):
            exclamation_count += 1
        case _:
            declarative_count += 1
```

```python
labels = ['Звичайні', 'Питальні', 'Окличні', 'Трикрапка']
values = [declarative_count, question_count, exclamation_count, ellipsis_count]
```

```python
plt.bar(labels, values, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Типи речень')
plt.ylabel('Кількість')
plt.title('Гістограма частоти появи різних типів речень у тексті')
plt.show()
```
